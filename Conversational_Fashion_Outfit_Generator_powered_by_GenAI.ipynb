```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
```

```python
!pip show torch
```

```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"

pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
prompt = """Muzzafarpur looking for a Karwa Chauth outfit to offer appropriate and versatile outfit
"""

image = pipe(prompt).images[0]
```

```python
print("[PROMPT]: ",prompt)
plt.imshow(image);
plt.axis('off');
```

```python
prompts = [
    "black suit 30year men for winter",
    "white suit 30year men for winter",
    "Red suit 30year men for Summer",
    "light blue 30year men for marriage"
]
fig = plt.figure(figsize=(16, 5))
# Generate and display images for each prompt
for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
    print(f"[PROMPT {i + 1}]: {prompt}")
    plt.subplot(1, len(prompts), i + 1)
    plt.imshow(image)
    plt.axis('off')

plt.show()
```
